```python
import pandas as pd
import os

# Read the CSV file into a DataFrame
file_path = 'mesh_eval_results.csv'
df = pd.read_csv(file_path)

# Edit pred_path to only show model/name (last two parts of the path)
if "pred_path" in df.columns:
    df["pred_path"] = df["pred_path"].apply(lambda x: os.path.join(*os.path.normpath(str(x)).split(os.sep)[-2:]))

# Drop unwanted columns
df = df.drop(columns=["gt_path", "plot_path", "compare_path"], errors="ignore")

# Set pandas display options for a nicer look
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)
pd.set_option('display.width', 1000)
pd.set_option('display.precision', 4)
pd.set_option('colheader_justify', 'center')

# Display the DataFrame with a styled header and alternating row colors
def style_dataframe(df):
    return df.style\
        .set_caption("Mesh Evaluation Results")\
        .set_table_styles(
            [{'selector': 'th', 'props': [('background-color', '#40466e'), ('color', 'white'), ('font-size', '110%')]}]
        )\
        .apply(lambda x: ['background-color: #f2f2f2' if i%2 else '' for i in range(len(x))], axis=0)

style_dataframe(df)
```
